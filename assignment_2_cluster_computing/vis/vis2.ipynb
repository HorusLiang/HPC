```python
import matplotlib.pyplot as plt
import numpy as np
x = np.asarray([31250000, 62500000, 125000000, 250000000, 500000000, 1000000000, 2000000000])
y1 = np.asarray([3.14159262158972163448,3.14159263758957907342,3.14159264558972584069,3.14159264958986250704,3.14159265159010692159,3.14159265259020425987,3.14159265308932411642])
y2 = np.asarray([0.187719,0.250580,0.435392,0.862533,1.705802,3.370263,6.679148 ])
```

```python
fig = plt.figure(figsize=(10,5))

color = 'purple'
ax = fig.add_subplot(111)
ax.plot(x, y2, marker='o', label = 'Execution time', color=color)
ax.legend(loc=0)
ax.grid()
ax.set_xlabel("niters")
ax.set_ylabel("Execution time (seconds)", color=color)
ax.set_title('Leibnitz PI Estimation with OpenMP')
ax.tick_params(axis='y', labelcolor=color)

color = 'darkgray'
ax2 = ax.twinx()
ax2.plot(x, y1, color=color, marker='.',ls=':', label = 'Estimated Pi')
ax2.set_ylabel("Estimated Pi",color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.axhline(np.pi, xmin=0.04, xmax=0.96, linewidth=10, color=color, alpha=0.3, label='Pi')
ax2.legend(loc=0)

plt.savefig('t2.png')
```
